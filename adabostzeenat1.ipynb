```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('Iris.csv')
#df
#df = df[(df['Species'] == 'Iris_versicolor') | (df['Species'] == 'Iris_virginica')]
print(df.head(5))
```

```python
df['probR1']=1/(df.shape[0])
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt1 = DecisionTreeClassifier(max_depth=1)
```

```python
X = df.iloc[:,1:5].values
y = df.iloc[:,5].values
```

```python
X
```

```python
y
```

```python
dt1.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(dt1)
```

```python
df['y_pred'] = dt1.predict(X)
```

```python
df
```

```python
#misclassified = 0 if the label and prediction are same
df.loc[df.Species != df.y_pred, 'predict'] = 1
df.loc[df.Species == df.y_pred, 'predict'] = 0
```

```python
er = sum(df['predict'] * df['probR1'])
```

```python
er
```

```python
def calculate_model_weight(error):

  return 0.5*np.log((1-error)/(error))
```

```python
# Step 3 - calculate model weight
alpha1 = calculate_model_weight(er)
alpha1
```

```python
# Step 4 - Update weights
def updateweights(row,alpha=0.423):
  if row['Species'] == row['y_pred']:
    return row['probR1'] * np.exp(-alpha)
  else:
    return row['probR1'] * np.exp(alpha)
```

```python
df['updated_weights'] = df.apply(updateweights,axis=1)
```

```python
df
```

```python
df['updated_weights'].sum()
```

```python
df['nomalized_weights'] = df['updated_weights']/df['updated_weights'].sum()
```

```python
df.tail(20)
```

```python
df['nomalized_weights'].sum()
```

```python
df['cumsum_upper'] = np.cumsum(df['nomalized_weights'])
```

```python
df['cumsum_lower'] = df['cumsum_upper'] - df['nomalized_weights']
```

```python
df[['SepalLength','SepalWidth','PetaiLength','PetalWidth','Species','probR1','y_pred','updated_weights','cumsum_lower','cumsum_upper']]
```

```python
def create_new_dataset(df):

  indices = []

  for i in range(df.shape[0]):
    a = np.random.random()
    for index,row in df.iterrows():
      if row['cumsum_upper'] > a and a > row['cumsum_lower']:
        indices.append(index)
  return indices
```

```python
index_values = create_new_dataset(df)

```

```python
second_df = df.iloc[index_values,[1,2,3,4,5,6]]
```

```python
second_df
```

```python
dt2 = DecisionTreeClassifier(max_depth=1)
```

```python
X = second_df.iloc[:,0:4].values
y = second_df.iloc[:,4].values
```

```python
y
dt2.fit(X,y)
```

```python
plot_tree(dt2)
```

```python
second_df['y_pred'] = dt2.predict(X)
```
